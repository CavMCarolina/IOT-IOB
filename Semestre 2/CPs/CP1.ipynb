# Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Estrela do mar

```python
img = cv2.imread("Estrela_do_Mar_2.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(img_rgb)
```

## Tentativas com Blur e Canny

```python
# imgb = cv2.blur(img_rgb, (10, 10))
# imgc = cv2.Canny(imgb, 0, 140)

imgb = cv2.blur(img_rgb, (9, 7))
imgc = cv2.Canny(imgb, 0, 190)

plt.imshow(imgc, cmap="gray")
```

## Contorno com HSV

```python
image_lower_hsv = np.array([0, 199, 132])
image_upper_hsv = np.array([7, 255, 232])

mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)

plt.imshow(mask_hsv, cmap="gray")
```

# Bolas de Basquete

```python
img = cv2.imread("basquete.jpg", 0)
img = cv2.medianBlur(img, 9)
cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,20,param1=50,param2=40,
                           minRadius=45,maxRadius=90)

if circles is not None:
  circles = np.uint16(np.around(circles))
  for i in circles[0,:]:
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)

plt.imshow(cimg)
```

# Tentativa com Balões

```python
img = cv2.imread("baloes.jpg", 0)
img = cv2.medianBlur(img, 9)
cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,20,param1=50,param2=30,
                           minRadius=30,maxRadius=90)

if circles is not None:
  circles = np.uint16(np.around(circles))
  for i in circles[0,:]:
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)

plt.imshow(cimg)
```
