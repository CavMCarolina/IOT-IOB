```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("bolinha.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)

image_lower_hsv = np.array([150, 50, 50])
image_upper_hsv = np.array([170, 255, 255])

mask = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
img2 = cv2.imread("monstros.jpg")
img_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img_hsv2 = cv2.cvtColor(img_rgb2, cv2.COLOR_RGB2HSV)

image_lower_hsv2 = np.array([30, 50, 50])
image_upper_hsv2 = np.array([79, 255, 255])

mask = cv2.inRange(img_hsv2, image_lower_hsv2, image_upper_hsv2)
plt.imshow(mask, cmap="gray")
plt.show()
```

```python

import ipywidgets as widgets
from ipywidgets import interact

# Carregar imagem
img = cv2.imread("monstros.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

def aplicar_hsv(hmin, smin, vmin, hmax, smax, vmax):
    # Intervalos
    lower = np.array([hmin, smin, vmin])
    upper = np.array([hmax, smax, vmax])

    # Criar mscara
    mask = cv2.inRange(img_hsv, lower, upper)
    result = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

    # Mostrar resultados
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.title("Original")
    plt.imshow(img_rgb)

    plt.subplot(1,3,2)
    plt.title("Mscara")
    plt.imshow(mask, cmap="gray")

    plt.subplot(1,3,3)
    plt.title("Filtrado")
    plt.imshow(result)
    plt.show()

# Sliders interativos
interact(
    aplicar_hsv,
    hmin=widgets.IntSlider(min=0, max=179, step=1, value=140),
    smin=widgets.IntSlider(min=0, max=255, step=1, value=50),
    vmin=widgets.IntSlider(min=0, max=255, step=1, value=50),
    hmax=widgets.IntSlider(min=0, max=179, step=1, value=170),
    smax=widgets.IntSlider(min=0, max=255, step=1, value=255),
    vmax=widgets.IntSlider(min=0, max=255, step=1, value=255),
)
```
